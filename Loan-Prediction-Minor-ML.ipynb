```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data Cleaning

```python
df.isnull()
```

```python
total = df.isnull().sum()
```

```python
total
```

```python
df['Gender'].value_counts()
```

```python
df.Gender = df.Gender.fillna('Male')
```

```python
df['Married'].value_counts()
```

```python
df.Married = df.Married.fillna('Yes')
```

```python
df['Dependents'].value_counts()
```

```python
df.Dependents = df.Dependents.fillna('0')
```

```python
df['Self_Employed'].value_counts()
```

```python
df.Self_Employed = df.Self_Employed.fillna('No')
```

```python
mean = df.LoanAmount.mean()
mean
```

```python
df.LoanAmount = df.LoanAmount.fillna(mean)
```

```python
df['Loan_Amount_Term'].value_counts()
```

```python
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(360.0)
```

```python
df['Credit_History'].value_counts()
```

```python
df.Credit_History = df.Credit_History.fillna(1.0)
```

```python
total = df.isnull().sum()
```

```python
total
```

## Variable into numeric

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

## Feature Selection

```python
x = df.iloc[:, 1:12]
y = df.iloc[:, 12]
```

```python
x
```

## Splitting

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y , test_size = 0.20, random_state = 0)
```

```python
x_test
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_predictions = model.predict(x_test)
```

```python
y_predictions
```

```python
y_test
```

## Model Evaluation

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_predictions)
cnf_matrix
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_predictions))
```

## Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()
logistic_regression.fit(x_train, y_train)
```

```python
y_pred = logistic_regression.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

## Confusion matrix

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sn.heatmap(cnf_matrix, annot=True)
plt.title('Confusion matrix')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
sn.heatmap(cnf_matrix/np.sum(cnf_matrix), annot=True, fmt='.2%')
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
